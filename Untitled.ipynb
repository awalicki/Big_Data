```python
query = """

select r20_21.*, stations.*

from (

select *
from `bigquery-public-data.noaa_gsod.gsod2020`

union all

select *
from `bigquery-public-data.noaa_gsod.gsod2021`

) as r20_21
join `bigquery-public-data.noaa_gsod.stations` as stations on r20_21.stn = stations.usaf and r20_21.wban = stations.wban
"""

query_job = client.query(query)
query_result = query_job.result()
# weather and stations
dfwas = query_result.to_dataframe()
```

```python
df_part6 = pd.read_csv('df_merged.csv')

df_energy = pd.read_csv('World Energy Consumption.csv')
df_co2 = pd.read_csv('annual-co-emissions-by-region.csv')
df_disasters = pd.read_csv('EmergencyEvents.csv', sep=';')

df_energy = df_energy[['country', 'year', 'renewables_electricity', 'fossil_electricity']].drop_duplicates()
df_energy.columns = ['country', 'year', 'renewables_twh', 'fossil_twh']

df_co2 = df_co2[['Entity', 'Year', 'Annual CO₂ emissions']].drop_duplicates()
df_co2.columns = ['country', 'year', 'co2_emissions_value']

df_disasters = df_disasters[['Country', 'Year', 'Total Events', 'Total Deaths']].copy()
df_disasters = df_disasters.groupby(['Country', 'Year']).agg(
    disasters_count=('Total Events', 'sum'),
    total_deaths=('Total Deaths', 'sum')
).reset_index()
df_disasters.columns = ['country', 'year', 'disasters_count', 'total_deaths']

df_energy = df_energy.dropna(subset=['country', 'year'])
df_co2 = df_co2.dropna(subset=['country', 'year'])
df_disasters = df_disasters.dropna(subset=['country', 'year'])
df_part6 = df_part6.dropna(subset=['Area', 'year'])

df_part6['year'] = df_part6['year'].astype(int)
df_energy['year'] = df_energy['year'].astype(int)
df_co2['year'] = df_co2['year'].astype(int)
df_disasters['year'] = df_disasters['year'].astype(int)

df_part6['merge_country'] = df_part6['Area'].astype(str).str.strip().str.lower()
df_energy['merge_country'] = df_energy['country'].astype(str).str.strip().str.lower()
df_co2['merge_country'] = df_co2['country'].astype(str).str.strip().str.lower()
df_disasters['merge_country'] = df_disasters['country'].astype(str).str.strip().str.lower()

kaggle_to_fao = {
    'united states': 'united states of america',
    'united states of america (the)': 'united states of america',
    'russia': 'russian federation',
    'russian federation (the)': 'russian federation',
    'turkey': 'türkiye',
    'china': 'china, mainland',
    'vietnam': 'viet nam',
    'south korea': 'republic of korea',
    'north korea': "democratic people's republic of korea",
    'bolivia': 'bolivia (plurinational state of)',
    'venezuela': 'venezuela (bolivarian republic of)',
    'iran': 'iran (islamic republic of)',
    'iran (islamic republic of)': 'iran (islamic republic of)',
    'taiwan (province of china)': 'china, taiwan province of',
    'democratic republic of congo': 'democratic republic of the congo',
    'cape verde': 'cabo verde',
    'micronesia (country)': 'micronesia (federated states of)',
    'micronesia': 'micronesia (federated states of)',
    "cote d'ivoire": "côte d'ivoire",
    "ivory coast": "côte d'ivoire",
    'laos': "lao people's democratic republic",
    'moldova': 'republic of moldova',
    'netherlands': 'netherlands (kingdom of the)',
    'syria': 'syrian arab republic',
    'east timor': 'timor-leste',
    'taiwan': 'china, taiwan province of',
    'tanzania': 'united republic of tanzania',
    'united kingdom': 'united kingdom of great britain and northern ireland'
}

df_energy['merge_country'] = df_energy['merge_country'].replace(kaggle_to_fao)
df_co2['merge_country'] = df_co2['merge_country'].replace(kaggle_to_fao)
df_disasters['merge_country'] = df_disasters['merge_country'].replace(kaggle_to_fao)

df_merged_energy = df_part6.merge(df_energy, on=['merge_country', 'year'], how='left', suffixes=('', '_y'))
df_merged_co2 = df_part6.merge(df_co2, on=['merge_country', 'year'], how='left', suffixes=('', '_y'))
df_merged_disasters = df_part6.merge(df_disasters, on=['merge_country', 'year'], how='left', suffixes=('', '_y'))

df_merged_energy.drop(columns=[col for col in df_merged_energy.columns if col.endswith('_y')] + ['merge_country'], inplace=True, errors='ignore')
df_merged_co2.drop(columns=[col for col in df_merged_co2.columns if col.endswith('_y')] + ['merge_country'], inplace=True, errors='ignore')
df_merged_disasters.drop(columns=[col for col in df_merged_disasters.columns if col.endswith('_y')] + ['merge_country'], inplace=True, errors='ignore')

df_merged_disasters['disasters_count'] = df_merged_disasters['disasters_count'].fillna(0)
df_merged_disasters['total_deaths'] = df_merged_disasters['total_deaths'].fillna(0)

df_merged_energy.to_csv('df_weather_energy.csv', index=False)
df_merged_co2.to_csv('df_weather_co2.csv', index=False)
df_merged_disasters.to_csv('df_weather_disasters.csv', index=False)
```
